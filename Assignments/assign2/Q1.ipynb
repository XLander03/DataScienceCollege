9)Analyse gender ratio of boys and girls among infants (age =0) 

```python
for i in range(6):
    boys = df.loc[[df['SEX'] == 1] & [df['AGE'] == 0], [f'201{i}']].values
    girls = df.loc[[df['SEX'] == 0] & [df['AGE'] == 0], [f'201{i}']].values
    ratio = boys[0][0]/girls[0][0]
    print(f"The ratio of boys to girls in the year 201{i} is {ratio}")

```
